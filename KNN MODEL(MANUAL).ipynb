```python
import pandas as pd 
import numpy as np

from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score
import matplotlib.pyplot as plt
```

### Load Dataset

```python
df = pd.read_csv("knn_telecom.csv")
```

### Encode Categorical Variables

```python
le_gender=LabelEncoder()
le_plan=LabelEncoder()

df["Gender"]=le_gender.fit_transform(df["Gender"])
df["PlanType"]=le_plan.fit_transform(df["PlanType"])
```

### Separate Features and Target

```python
X=df.drop("Churn",axis=1)
y=df["Churn"]
```

```python
### Feature Scaling(Required for KNN)
```

```python
scaler = StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
### Cross Validation Setup
```

```python
sfk=StratifiedKFold(n_splits=3, shuffle=True,random_state=42)
k_values =range(1,8)
cv_scores=[]
```

```python
### Train and Evaluate for Different K Values
```

```python
for k in k_values: 
    knn =KNeighborsClassifier(n_neighbors=k)
    scores=cross_val_score(knn, X_scaled, y, cv=sfk, scoring="accuracy")
    cv_scores.append(scores.mean())
    
```

```python
### Display Result
```

```python
for k,score in zip(k_values,cv_scores):
    print("K=",k," |Cross Validation Accuracy =", round(score, 3))
    
```

```python
### Select Best K
```

```python
best_k=k_values[np.argmax(cv_scores)]
print("\nBest K: ",best_k)
```

```python
### Plot Result
```

```python
plt.plot(k_values,cv_scores)
plt.xlabel("K Value")
plt.ylabel("Cross Validation Acuracy")
plt.title("Finding Optimal K")
plt.show()
```
